### A: Experiment with Cross-filtering

The visualization widgets in Databricks dashboards offer interactivity by default. Let's see this capability in action.

**📌 NOTE:** Many of the following instructions take place in a non-notebook area of the platform. It is recommended to open either the notebooks in a separate tab or window to reference the instructions alongside the area where they are performed.

1. Select **Dashboards** from the sidebar navigation. (It is recommended to right-click and select Open Link in New Window or Tab)

1. Click on the **Canada Sales** dashboard you began building in an earlier activity.
1. Navigate to the **Overview** page tab.
1. Click on one of the segments of the pie chart. What happens to the rest of the display? Do all of the visualizations react, or just some of them?
1. Click again on the same segment of the pie chart to restore the display to what it was previously.
1. Experiment with clicking on other graphical components to see the reactions. Don't forget to undo your actions by clicking again. How can you tell that cross-filtering is in action? 

Cross-filtering applies to all the visualizations on a given page of a dashboard that are based on the same dataset.

### B: Implement a Local Filter Widget

Filter widgets affect all visualizations on the same page that are based on the same dataset. They offer finer grained control that does cross-filtering, and they also make your dashboard more self-describing. Anyone who sees the filter widget's current settings will assume that the other dashboard components they see are being filtered. 

1. If you moved away, select **Dashboards** from the sidebar navigation.
1. Click on the **Canada Sales** dashboard you began building in an earlier activity.
1. Navigate to the **Overview** page tab.
1. Click the icon in the palette for adding a filter, which looks like a tiny funnel.
1. Land the new widget wherever it is convenient. Click in the new widget to give it focus.
1. When the filter widget is selected, the filter configuration panel appears on the right side of the screen.

1. Apply the following settings:  
  - **Filter**: Date range picker
  - **Fields**: Expand the **Orders22** dataset and choose **orderdate**
8. Use the checkboxes to turn on **Title**.  

1. Click the placeholder title on the new filter widget and change it to **Order Date**  
1. Experiment with your filter widget. Which visualizations change when you define a date range? Do the tables on the other page change?

### C: Implement a Global Filter

Global filters affect all visualizations throughout the dashboard that are based on the same dataset. They offer more global control of filtering. 

1. If you moved away, select **Dashboards** from the sidebar navigation.

1. Click on the **Canada Sales** dashboard you began building in an earlier activity.
1. Between the **Data** tab and the **Overview** tab, locate the Filter icon, which looks like a tiny funnel, and click it. This opens the **Global filters** side panel.
1. Click the **+** icon at the upper right of the side panel. 
1. In the Widget configuration panel, set the following:
  - **Title:** Province Filter
  - **Filter:** Single Value
  - **Fields:** 
    - Table: Orders22
    - Field: province
