    - ca_products

These tables contain simulated business-to-business order and opportunity data for an imaginary company's Canadian sales operation. The **ca_orders** table contains information from 2021 through the middle of November 2024. This table identifies the relevant customer with a unique key that points into the **ca_customers** table, and it identifies the relevant product with a unique key that points into the **ca_products** table. The **ca_opportunities** table provides additional details regarding sales opportunities and sales representatives. You'll use the data in these tables to prepare your dashboard.

### A: Define a Dataset with AI Help

In this exercise we'll define a new dashboard and let Databricks Assistant help us compose the SQL query that will define the dataset.

**📌 NOTE:** Throughout these instructions, replace YOURSCHEMA with the name of your schema. (You can find your schema name in the setup run cell that you have just executed.) Additionally, many of the following instructions take place in a non-notebook area of the platform. It is recommended to open either the notebooks in a separate tab or window to reference the instructions alongside the area where they are performed.

1. Select **Dashboards** from the sidebar navigation. (It is recommended to right-click and select Open Link in New Window or Tab)
1. Click the **Create dashboard** button at top right.
1. At the top left of the dashboard definition panel, you'll see the placeholder dashboard name, which will be something like `New Dashboard 202X-01-01 12:00:00`. Click on it and change its name to **Orders and Opportunities 2023**.
1. In the resulting dashboard-creation screen, click the **Data** tab. 
1. You will find three choices: **Create from SQL**, **Add data source**, and **Upload file**. Choose the other option, **Create from SQL**.
1. Click on the Databricks Assistant icon on the left side of the dashboard-definition panel (a small four-pointed star). The Databricks Assistant chat window opens up at right. 

    **📌 NOTE:** Note the Assistant may give different results. Please read carefully and confirm the correct query below.

1. Enter this prompt: (Remember to replace YOURSCHEMA with your unique schema name.)
<br/>

    ```
    Give me a query with all the fields in the
    dbacademy.YOURSCHEMA.ca_orders table, joined with the dbacademy.YOURSCHEMA.ca_customers
    ```

  In the chat window at right, Databricks Assistant proposes a query.

8. Let's refine this query. In the chat text-entry field below that suggested query, enter this prompt:
<br/>

    ```
    Leave out the productID field
    ```

1. One more refinement. Enter this prompt:
<br/>

    ```
    I only want to see orders in calendar year 2023
    ```

1. Confirm the query shows the following. If not simply use the query provided below.
<br/>

    ```SQL
    SELECT o.orderid,
          o.orderdate,
          o.quantity,
          o.orderamt,
          o.salesrep,
          c.customerid,
          c.customername,
          c.city,
          c.province
    FROM dbacademy.YOURSCHEMA.ca_orders o
    JOIN dbacademy.YOURSCHEMA.ca_customers c
      ON o.customerid = c.customerid
    WHERE YEAR(o.orderdate) = 2023
    ```

11. When you are happy with the query, click the small "Copy" icon in Databricks Assistant's response window and paste the SQL query into the dataset definition dialogue's main screen. Click **Run** to confirm that the query works. Feel free to adjust it if necessary, either manually or by clicking Databricks Assistant's help button to diagnose the problem and suggest a repaired query.

12. Click on the kebab menu to the right of the working name of this dataset ("Untitled dataset") and rename the dataset to **orders2023**.



### B: Define a Visualization with AI Help

In this exercise we'll let Databricks Assistant help us build a visualization based on our dataset.

1. Click on the **Untitled page** canvas tab at top left to switch to the canvas view.
1. At the bottom of the screen you have a palette for adding various widgets. Click the icon in the palette for adding a visualization and drag the resulting widget to the top of your canvas.

1. Click into the just-placed widget. Notice that Databricks Assistant offers a box into which you can type a text prompt. Enter this prompt:
<br/>

      ```
      bar chart of Total Sales by order month
      ```
