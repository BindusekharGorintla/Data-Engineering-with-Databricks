## B. Upload a Local CSV File to a Volume Using the UI

In this section, we will demonstrate how to upload a local CSV file to a Databricks volume. To replicate this process, we have provided the CSV file in the **data** folder for you to download to your local computer.

**Note:** Depending on your laptopâ€™s security policies, you might not be able to download and upload a file to Databricks.

1. Complete the following steps to download the CSV file locally:

   a. In the **M02 - Data Importing** folder, there is a **data** folder. Open the **data** folder, and you should see a file named **au_products.csv**.

   b. Hover your mouse over the **au_products.csv** file, then select the three ellipses to the right of the file and choose **Download**. This will download the CSV file to your computer's download folder.

2. Complete the following steps to upload the CSV file to your volume:

   a. In the far-left pane, right-click **Catalog** and select *Open in New Tab*.

   b. In the Catalog Explorer, select the **dbacademy** catalog.

   c. Select your **labuser** schema.

   d. Select **Volumes**, then choose the **myfiles** volume.

   e. In the upper-right corner, select **Upload to this volume**.

   f. Upload the **au_products.csv** file you downloaded in the previous step.

   g. Select **Upload** to upload the CSV file to the Databricks volume.

   h. Close the Catalog Explorer.

   **Note:** If you have the necessary permissions on your laptop to upload a file, the file will be uploaded to the **myfiles** volume.

3. Complete the following steps to view the CSV file in the Databricks volume:

   a. In the left navigation pane, select the catalog icon ![Catalog Icon](../Includes/images/catalog_icon.png).

   b. Expand the **dbacademy** catalog.

   c. Expand your **labuser** schema.

   d. Expand **Volumes** and your **myfiles** volume. You should see that the **au_products.csv** file has been uploaded to the Databricks volume (you might have to refresh the volume using the refresh button above the catalog search bar).

   **Note:** Remember, Databricks volumes are logical storage spaces used to store and access files. They are part of Unity Catalog, which is used to govern non-tabular datasets like CSV files.

4. With the CSV file in a volume, you can now programmatically display and create a table from the CSV file. We will show you how to do this in the next demonstration.

## C. Upload a Local CSV File as a Table Using the UI
In this section we will upload the same **au_products.csv** file to Databricks, but this time create a table using the UI instead of uploading the CSV file directly to a volume.

1. Complete the following steps to upload the CSV file as a table to your schema:

   a. In the far-left pane, right-click **Catalog** and select *Open in New Tab*.

   b. In the Catalog Explorer, select the **dbacademy** catalog.

   c. Select your **labuser** schema.

   d. In the upper-right corner, select the **Create** dropdown.

   e. Select **Table**.

   f. Add your locally downloaded file, **au_products.csv**, to the UI.

   g. Notice that the **Create or modify table from file upload** provides a variety of options to set when importing your CSV file as a table.

   h. Change the name of the table to **aus_products_ui**.

   i. Since this is a simple CSV file, leave the remaining options as they are and select **Create table** at the bottom right. This will create the table in Databricks from your local CSV file.

   j. After the table is created, accept the **AI Suggested Description**.

   **Note:** If you have the necessary permissions on your laptop to upload a file, the file will be uploaded as a new table named **aus_products_ui** to Databricks. However, your security policies might prohibit uploading a file.


2. In the left navigation pane refresh your **labuser** schema. The new table **aus_products_ui** should appear in your schema.
