## C. Unified SQL Editor
Throughout this course, we've been using an SQL notebook to execute DBSQL. Using the notebook allows us to incorporate markdown text during training.

However, as a developer, you might prefer to use a native SQL editor, especially if you have experience with one in another SQL development environment. In Databricks, you can use the Unified SQL Editor for a more developer focused experience with specific features.

### C1. Copy the Unified SQL Editor Instructions

1. In the left navigation right click on the file **3.1b - Unified SQL Editor Directions.txt** and select **Open in new tab**. Then copy the contents of the file to your clipboard.

### C2. Opening and Using the Unified SQL Editor

1. In the far left main navigation pane right click on **SQL Editor** and select **Open in New Tab**. This will take you directly to the editor.

2. Paste the directions from step **C1** that you copied into the query editor. Then follow those instructions.
